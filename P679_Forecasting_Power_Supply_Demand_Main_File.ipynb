The PJM power consumption series is stationary.

Since the p-value is less than 0.05, the null hypothesis of non-stationarity is rejected. Therefore, the power consumption time series is stationary and suitable for time-series forecasting models.

```python
# Train-Test Split (80% Train, 20% Test)

train_size = int(len(df) * 0.8)

train = df.iloc[:train_size]
test = df.iloc[train_size:]

print("Training Data Shape:", train.shape)
print("Testing Data Shape:", test.shape)
```

```python
print("Train Start:", train.index.min())
print("Train End:", train.index.max())

print("Test Start:", test.index.min())
print("Test End:", test.index.max())

# This ensures:

# Training data contains older observations.
# Testing data contains future observations.
# No data leakage occurs.
```

# **1. ARIMA**

```python
# Train Arima
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(train['PJMW_MW'], order=(2,0,2))
model_fit = model.fit()

print(model_fit.summary())
```

```python
# Forecast on Test Data
forecast = model_fit.forecast(steps=len(test))
```

```python
# Evaluate the Model
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(test['PJMW_MW'], forecast)
rmse = np.sqrt(mean_squared_error(test['PJMW_MW'], forecast))

print("MAE:", mae)
print("RMSE:", rmse)
```

On average, predictions are off by about 808 MW.

Larger errors are penalized more heavily in RMSE, which is about 997 MW.

Since PJM power consumption values are typically in the range of several thousand MW, an error around 800–1000 MW is reasonable for a first ARIMA model. **bold text**

```python
# Visualize Actual vs Predicted

plt.figure(figsize=(15,5))
plt.plot(test.index, test['PJMW_MW'], label='Actual')
plt.plot(test.index, forecast, label='Forecast')
plt.title('ARIMA Forecast vs Actual')
plt.xlabel('Date')
plt.ylabel('Power Consumption (MW)')
plt.legend()
plt.show()
```

```python
# Check Residuals
residuals = model_fit.resid

plt.figure(figsize=(12,4))
plt.plot(residuals)
plt.title("Residuals")
plt.show()
```

# **2. SARIMA**

```python
train_small = train.iloc[-10000:]
```

```python
print(train_small.shape)
print(train_small.columns)
```

```python
pjm_series = train_small['PJMW_MW']

print(pjm_series.shape)
print(type(pjm_series))
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

pjm_series = train.iloc[-10000:]['PJMW_MW']

sarima_model = SARIMAX(
    pjm_series,
    order=(1,0,1),
    seasonal_order=(1,0,1,24)
)

sarima_fit = sarima_model.fit(disp=False)
```

```python
test_small = test.iloc[:1000]
```

```python
sarima_forecast = sarima_fit.forecast(steps=len(test_small))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(test_small['PJMW_MW'], sarima_forecast)
rmse = np.sqrt(mean_squared_error(test_small['PJMW_MW'], sarima_forecast))

print("SARIMA MAE:", mae)
print("SARIMA RMSE:", rmse)
print("AIC:", sarima_fit.aic)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))
plt.plot(test_small.index, test_small['PJMW_MW'], label='Actual')
plt.plot(test_small.index, sarima_forecast, label='SARIMA Forecast')
plt.legend()
plt.show()
```

**The SARIMA model reduced forecasting error by approximately 26% (MAE) and 19% (RMSE) compared to ARIMA.**

# **3. Holt-Winters**

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
train_hw = train.iloc[-10000:]['PJMW_MW']
test_small = test.iloc[:1000]
```

```python
hw_model = ExponentialSmoothing(
    train_hw,
    trend='add',
    seasonal='add',
    seasonal_periods=24
)

hw_fit = hw_model.fit()
```

```python
hw_forecast = hw_fit.forecast(steps=len(test_small))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(test_small['PJMW_MW'], hw_forecast)
rmse = np.sqrt(mean_squared_error(test_small['PJMW_MW'], hw_forecast))

print("Holt-Winters MAE:", mae)
print("Holt-Winters RMSE:", rmse)
```

```python
plt.figure(figsize=(15,5))
plt.plot(test_small.index, test_small['PJMW_MW'], label='Actual')
plt.plot(test_small.index, hw_forecast, label='Holt-Winters Forecast')

plt.title('Holt-Winters Forecast vs Actual')
plt.xlabel('Date')
plt.ylabel('PJMW_MW')
plt.legend()
plt.show()
```

**Holt-Winters achieved the lowest MAE and RMSE.**

**It outperformed SARIMA by:**

**MAE improvement ≈ 91 MW**

**RMSE improvement ≈ 150 MW**

**This indicates that the PJMW electricity demand series has a strong and relatively stable trend + seasonal pattern that Holt-Winters captures very effectively.**

```python
prophet_df = df.reset_index()[['Datetime', 'PJMW_MW']]

prophet_df.columns = ['ds', 'y']

prophet_df.head()
```

```python
print(df.index.name)
```

```python
train_size = int(len(prophet_df) * 0.8)

train_prophet = prophet_df.iloc[:train_size]
test_prophet = prophet_df.iloc[train_size:]
```

```python
train_prophet = train_prophet.iloc[-10000:]
test_prophet = test_prophet.iloc[:1000]
```

```python
from prophet import Prophet

model = Prophet(
    daily_seasonality=True,
    weekly_seasonality=True,
    yearly_seasonality=True
)

model.fit(train_prophet)
```

```python
future = model.make_future_dataframe(
    periods=len(test_prophet),
    freq='h'
)

forecast = model.predict(future)
```

```python
prophet_forecast = forecast['yhat'].tail(len(test_prophet)).values
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(
    test_prophet['y'],
    prophet_forecast
)

rmse = np.sqrt(
    mean_squared_error(
        test_prophet['y'],
        prophet_forecast
    )
)

print("Prophet MAE:", mae)
print("Prophet RMSE:", rmse)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))

plt.plot(
    test_prophet['ds'],
    test_prophet['y'],
    label='Actual'
)

plt.plot(
    test_prophet['ds'],
    prophet_forecast,
    label='Prophet Forecast'
)

plt.legend()
plt.title('Prophet Forecast vs Actual')
plt.show()
```

**Compared to ARIMA:**

**MAE reduced from 808.14 → 444.76 (45% improvement)**
**RMSE reduced from 996.51 → 547.13 (45% improvement)**

**Compared to Holt-Winters:**

**MAE reduced from 502.92 → 444.76**
**RMSE reduced from 660.74 → 547.13**
